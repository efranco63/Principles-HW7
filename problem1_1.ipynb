```python
pd.set_option('display.max_columns',500)

three11 = pd.read_csv('311_data.csv')
zips = pd.read_csv('zipCodePopulationData.csv')

heads = ['Agency','Incident Zip']
three11 = three11[heads]
three11 = three11[pd.notnull(three11['Incident Zip'])]
three11 = three11.reset_index()
three11 = three11.drop('index',1)

zips311 = three11.groupby('Incident Zip').aggregate('count')
zips311_list = zips311.index.tolist()
zips311_counts = zips311['Agency'].tolist()
for i in xrange(0,len(zips311_list)):
    if type(zips311_list[i]) == str:
        if '-' in zips311_list[i]:
            x = zips311_list[i].index('-')
            zips311_list[i] = zips311_list[i][:x]
zips311_list = zips311_list[:-4]
zips311_counts = zips311_counts[:-4]
for i in xrange(0,len(zips311_list)):
    zips311_list[i] = int(zips311_list[i])

zip_counts = {}
for i in xrange(0,len(zips311_list)):
    if zips311_list[i] in zip_counts:
        zip_counts[zips311_list[i]][0] += zips311_counts[i]
    else:
        zip_counts[zips311_list[i]] = []
        zip_counts[zips311_list[i]].append(zips311_counts[i])

for i in xrange(0,len(zips['Zip Code ZCTA'])):
    if zips['Zip Code ZCTA'][i] in zip_counts:
        zip_counts[zips['Zip Code ZCTA'][i]].append(zips['2010 Census Population'][i])

zip_pop = []
zip_comp = []
for i in zip_counts:
    if len(zip_counts[i]) == 2:
        zip_comp.append(zip_counts[i][0])
        zip_pop.append(zip_counts[i][1])

        
plt.scatter(zip_pop,zip_comp)
plt.title('Number of Complaints by Population by Zip Code')
plt.ylabel('Number of Complaints')
plt.xlabel('Population')
plt.xlim(-5000,)
plt.ylim(-500,)
plt.show()
```

```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
%matplotlib inline

with open('311_data.csv') as f:
    data = csv.DictReader(f)
    zips = []
    Agents = []
    A_list = ['NYPD','DOT','DOB','TLC','DPR']
    for line in data:
        if line['Agency'] in A_list:
            zipcode = line['Incident Zip']
            agent = line['Agency']
            if '-' in zipcode: 
                x = zipcode.index('-')
                zipcode = zipcode[:x]
            if zipcode.isdigit():
                zips.append(int(zipcode))
                Agents.append(agent)
        
zips_df = pd.DataFrame()
zips_df['zips'] = zips
zips_df['agency'] = Agents
zips_df['counts'] = 1

zipgroups = zips_df.groupby(['zips','agency']).aggregate(sum)
test = zipgroups.unstack().fillna(0)
A_list = ['DOB','DOT','DPR','NYPD','TLC']

zip_list = []
count_list = []
agent_list = []

for i in test.iterrows():
    index,data = i
    x = data.tolist()
    #append the zip code
    zip_list.append(index)
    #append the max count
    count_list.append(max(x))
    #append the agency corresponding to max count
    agent_list.append(A_list[x.index(max(x))])
    
df = pd.DataFrame()
df['zips'] = zip_list
df['counts'] = count_list
df['agencies'] = agent_list

df = df.set_index('zips')

zip_pop = pd.read_csv('zipCodePopulationData.csv')

zip_pop = zip_pop.set_index('Zip Code ZCTA').astype(int)
df = df.join(zip_pop)

df = df[pd.notnull(df['2010 Census Population'])]   
    
temp_df = df[df['agencies']=='DOB']
DOB = plt.scatter(temp_df['2010 Census Population'], temp_df['counts'], color='Red')
temp_df = df[df['agencies']=='DOT']
DOT = plt.scatter(temp_df['2010 Census Population'], temp_df['counts'], color='DarkOrange')
temp_df = df[df['agencies']=='DPR']
DPR = plt.scatter(temp_df['2010 Census Population'], temp_df['counts'], color='Blue')
temp_df = df[df['agencies']=='NYPD']
NYPD = plt.scatter(temp_df['2010 Census Population'], temp_df['counts'], color='Green')
temp_df = df[df['agencies']=='TLC']
TLC = plt.scatter(temp_df['2010 Census Population'], temp_df['counts'], color='BlueViolet')
    
plt.legend((DOB,DOT,DPR,NYPD,TLC),
           ('DOB', 'DOT', 'DPR', 'NYPD', 'TLC'),
           ncol=2,
           scatterpoints=1,
           fontsize=8,
           loc='upper left')
    
    
# plt.scatter(df['2010 Census Population'],df['counts'],c=df['agencies'])
plt.title('Number of Complaints by Population by Zip Code')
plt.ylabel('Number of Complaints')
plt.xlabel('Population')
plt.xlim(-5000,)
plt.ylim(-20,)
plt.savefig('tester.png')
plt.show()
```
